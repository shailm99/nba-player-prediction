```python
import pandas as pd
import numpy as np
import os
```

```python
paths = os.listdir('../data/features/')
```

```python
import re
master_dataframe = pd.read_csv(f'../data/features/{paths[0]}')
year_end = re.search('\_(\d+)', paths[0]).group()[1:]
master_dataframe['year_end'] = year_end
master_dataframe
```

```python
for path in paths[1:]:
    df = pd.read_csv(f'../data/features/{path}')
    year_end = re.search('\_(\d+)', path).group()[1:]
    df['year_end'] = year_end
    master_dataframe = pd.concat([master_dataframe, df])
```

```python
master_dataframe.year_end.value_counts().sort_index()
```

```python
targets = pd.read_csv(f'../data/epm values 14-22.csv')
targets
```

```python
# we wont be working with 2021-2022 in training our model since this will be for future predictions for the 2022-23 season
master_dataframe = master_dataframe[master_dataframe.year_end != '2022']
master_dataframe.year_end.value_counts().sort_index()
```

```python
targets['feature_season'] = targets['season'] - 1
```

```python
targets
```

```python
master_dataframe['year_end'] = master_dataframe['year_end'].astype('int')
```

```python
combined_df = master_dataframe.merge(targets, how='inner', right_on = ['feature_season', 'player_name'], 
left_on= ['year_end', 'name'], suffixes=('_feature', '_target'))
```
